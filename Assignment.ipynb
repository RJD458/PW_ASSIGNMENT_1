● What is Data Science?

**Answer** :     
- Data Science : Data science is an interdisciplinary field from where we can determine meaningful knowledge, trends and actionable insights from raw datasets. In simple word Data Science is collecting, organizing, analyzing and interpreting data to solve real world problems and decision making.
Data Science is made up of several disciplinary fields, like -




1.   Mathematics and Statistics,
2.   Programming and Coding,
3.   Domain Knowledge,
4.   Machine Learning and Artificial Intelligence,
5.   Data Visualization

All these fields are essential to implement data science to solve real world problems.





- **More than Statistics** : In statistics we can only find the trends and patterns of the data but in data science we also do model development, dta visualization, handle big datasets. Statistics works on small datasets not big datasets. So, statistics is a part of Data Science not more than Data Science. Statistics is the foundation of Data Science.



- **More than Programming** : With the help of programming we clear datasets, we visualize data and build Machine Learning models but without the knowledge of statistics and domain knowledge it was meaningless. With the help of programming we handle big datasets. So, programming is a toolkit for Data Science that is why Data Science is more than progamming.

    - **Role of Statistics** : In this problem we can only find the relationship between several factors like blood pressure, suger level, heart beat etc. and diseases. We find the trade and patterns only with the help of statistics.


    - **Role of Programming** : With the help of programming we clean data  sets , visualize dtatsets and build a predictive model but without statistics and domain knowledge it is not possible.


  But Data Science works on both of them and beyond of them. It is a integrated model of several fields.

Question 4: An e-commerce company wants an AI system that:

1. Why this is an example of Agentic AI.
2. What risks might arise?
3. What human oversight would still be required?

**Answer** :
- Agentic AI refers to AI systems that can perceive, reason, decide, and act autonomously toward a goal.E-commerce company is fit for Agentic AI because -
  1. **Monitor Inventory** : The AI system continuously tracks stock level of the product.This is the perception stage of the AI system.
  
  2. **Predicts Demand** : Using historical sales, seasonal trends, and market behavior, the AI forecasts future product demand.This is the reasoning and prediction stage.

  3. **Automatically places Suppliers orders** : Based on predicted demand and current stock, the AI decides when and how much to reorder.It then places purchase orders without waiting for human approval.This is the decision-making and execution stage.

  4. **Negotiates price via email** : The AI communicates with suppliers, compares offers, and negotiates better pricing.This is a complex interaction stage, requiring contextual understanding and strategic responses.

  That complex thinking and behaviour and move towards the goal made the solution to Agentic AI.

- **Risks that might arise** :
  - **Wrong Decision Due to poor data** : If sales data is incomplete or inaccurate, demand predictions may fail.
  - **Uncontrolled Spending** : If the AI places large supplier orders without proper limits, the company may overspend.
  - **Security** : If external actors manipulate supplier emails or system inputs, the AI could:

    - send confidential data
    - approve fraudulent orders
    - engage with fake suppliers

    This creates cybersecurity concerns.

- **Human Oversights** :
  - **Monitoring** : Managers must regularly review:

      - AI decisions
      - order histories
      - supplier communications
      - system performance

    This ensures errors are detected early.
  - **Improvement Checking** : AI models must be retrained and updated.Human experts should validate predictions and refine strategies over time.
  - **Complain Elaboration** : To elabprate the complains of the customer to analize their sentiment properly.

```python
'''
Question 7: You are given a customer review:
review = "The product quality is Excellent but delivery was Late"
Write Python code to:
1. Convert to lowercase
2. Count number of words
3. Check if the word "excellent" exists
4. Replace "Late" with "delayed"

'''

review = "The product quality is Excellent but delivery was Late"

print("The lowercase version of review is : '" + review.lower() + "'") # Part 1

print("The number of words in review is : ", len(review.split())) # Part 2

# Part 3

bool = "excellent" in review
if bool:
  print("The word 'excellent' exists in review.")
else:
  print("The word 'excellent' does not exist in review.")

# Part 4

print("After replacing 'Late' with 'delayed' the string review is : '" + review.replace("Late","delayed") + "'")
```

```python
'''
Question 8: Given sales data:
sales = [12000, 15000, 11000, 18000, 17000]
1. Calculate total revenue.
2. Find average revenue.
3. Identify highest sales month.
4. Append new month sale = 20000.
5. Sort the list in descending order.

'''

sales = [12000, 15000, 11000, 18000, 17000]

# Part 1

total_revenue = 0
for i in sales:
  total_revenue += i
print("The total revenue is : ", total_revenue)

# Part 2

print("The average revenue is : ", total_revenue/len(sales))

# Parth 3

highest_sales_month = 0
for i in range(len(sales)):
  if(sales[highest_sales_month] < sales[i]):
    highest_sales_month = i
print("Highest sales month is : ", highest_sales_month+1)

# Part 4

sales.append(20000)
print("Sales after appending a new month sale = 20000 is : ", sales)

# Part 5

sales.sort(reverse=True)
print("Sales after sorting in descending order is : ", sales)
```
